```python
def prepare(df):
    df=df.transpose()
    df.columns=df.iloc[0:3].fillna('').apply(' '.join).str.strip()
    df=df.iloc[3:,1:]
    df.reset_index(drop=True, inplace=True)
    #df.replace('BAD',np.nan,inplace=True)
    #df.replace('#ERROR : 33329 : Invalid point number, the point may not exist on the server.',np.nan,inplace=True)
    #df.replace('#ERROR : 33330 : Could not resolve parameter name to a number.',np.nan,inplace=True)
    #df.replace('#ERROR : 33444 : Unknown MessageName : err33444',np.nan,inplace=True)
    #df=df.astype(float)
    #df= df.T.drop_duplicates().T
    return df
```

```python
hrr=['00:00:00', '01:00:00', '02:00:00','03:00:00','04:00:00','05:00:00', '06:00:00','07:00:00','08:00:00','09:00:00','10:00:00',
'11:00:00','12:00:00','13:00:00','14:00:00', '15:00:00','16:00:00','17:00:00','18:00:00','19:00:00','20:00:00','21:00:00'
,'22:00:00','23:00:00']
```

```python
# Unit_02
pattern = '(Unit02 ).*(xls|xlsx)' #.* means any number of characters between the fist group and second group
os.chdir(root_path)
df_macro_dict={}
for path,folders,files in os.walk(root_path):
    for file in files:
        if re.search(pattern,file):
            try:
                os.chdir(path)
                today= file.split('.')[0].split( )[1]
                day=int(today.split('_')[0])
                month=int(today.split('_')[1])
                year = int(today.split('_')[2])
                df_macro_dict[file]=pd.read_excel(file, header = 7, usecols='a:c,e:ab')#,skiprows=[15,34,49,50,66,67,127])
                df_macro_dict[file]=prepare(df_macro_dict[file])
                df_macro_dict[file].insert(0, column='hours', value=np.nan)
                df_macro_dict[file]['hours']=hrr
                df_macro_dict[file].insert(0, column='ddtt', value=datetime(int(year), int(month), int(day))) # creat column at the begiining
                df_macro_dict[file].insert(0, column='Date', value=np.nan)
                df_macro_dict[file]['Date']= df_macro_dict[file]['ddtt'].astype(str)+ ' '+df_macro_dict[file]['hours'].astype(str)
                df_macro_dict[file]['Date'] = pd.to_datetime(df_macro_dict[file]['Date'])
                df_macro_dict[file]=df_macro_dict[file].drop(['ddtt','hours'],axis=1)
                df_macro_dict[file]=df_macro_dict[file].reset_index(drop=True)
                df_macro_dict[file]= df_macro_dict[file].loc[:,~df_macro_dict[file].columns.duplicated()]
            except:
                print(f' file {file} could not be processed')
                continue
df_macro=pd.concat(df_macro_dict.values(),axis=0,ignore_index=True)
df_all['NHT_macro']=df_macro.copy()
```
